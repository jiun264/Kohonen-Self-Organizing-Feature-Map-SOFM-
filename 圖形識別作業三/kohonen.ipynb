```python
class kohonen:
    
    def __init__(self):
        self.W=[]
        self.M=0
        self.kohonenamount=0
        self.feature_dim=0
        self.output=[]       
        
    def initial(self,inputamount,kohonenamount):
        # 设置神经元节点数量
        self.M = inputamount  #輸入層神经元节点数量
        self.kohonenamount=kohonenamount #kohonen layer neuron amount
        # 获取输入数据的特征维度
        self.feature_dim = 13
        # 设置权重的范围
        weight_range = 0.1
        # 随机初始化权重矩阵
        self.W = np.random.uniform(-weight_range, weight_range, size=(self.M, kohonenamount))


    def forward(self, data):
        max_value = -np.inf
        max_index = -1
        W = []
        for i in range(self.feature_dim):
            input_sum = np.sum(data[i] * self.W[i])
            if input_sum > max_value:
                max_value = input_sum
                max_index = i
                W.append(self.W[i])

        return max_value, max_index


    
    def backward(self, max_value, max_index, data, learning_rate):
        delta_W = np.zeros_like(self.W[max_index])
        for i in range(len(self.W[max_index])):
            delta_W[i] = learning_rate * (data[i] - self.W[max_index][i])
        self.W[max_index] += delta_W

```

```python
feature_dim = 13
kohonen_layer = 5
k1 = kohonen()
k1.initial(feature_dim, kohonen_layer)

for index, row in data.iterrows():
    max_value, max_index = k1.forward(row.values)
    k1.backward(max_value, max_index, row.values, 0.1)

print(k1.W)

```
